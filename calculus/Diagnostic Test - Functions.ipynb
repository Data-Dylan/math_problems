### Question 2

$$
f(x) = \frac{f(2 + h) - f(2)}{h} \\
\Rightarrow f(x) = \frac{(2 + h)^3 - (2)^3}{h} \\
\Rightarrow f(x) = \frac{(h^2 + 4h + 4)(2 + h) - 8}{h} \\
\Rightarrow f(x) = \frac{h^3 + 4h^2 + 4h + 2h^2 + 8h + 8 - 8}{h} \\
\Rightarrow f(x) = \frac{h^3 + 6h^2 + 12h}{h} \\
\Rightarrow f(x) = h^2 + 6h + 12
$$

### Question 6
**(a)**

$$
f(-2) = 1 - x^2 \\
\Rightarrow f(-2) = 1 - (-2)^2 \\
\Rightarrow f(-2) = 1 - 4 \\
\Rightarrow f(-2) = -3 
$$
<br>
$$
f(1) = 2x + 1 \\
\Rightarrow f(1) = 2(1) + 1 \\
\Rightarrow f(1) = 2 + 1 \\
\Rightarrow f(1) = 3
$$

**(b)**

```python
# Create the piecewise function.
f = Piecewise((1 - x ** 2, x <= 0),
              (2 * x + 1, x > 0))

# Plot the function.
plot(f,
     xlim = (-5, 5),
     ylim = (-5, 5));
```
