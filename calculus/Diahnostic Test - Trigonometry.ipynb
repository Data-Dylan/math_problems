```python
# Software packages.
import math
import numpy as np
from matplotlib import pyplot as plt

# Function for showing triangles in a unit circle.
def unit_circle(xy):

    # Create figure and axis.
    fig, ax = plt.subplots()

    # Add a unit circle to the plot.
    circ = plt.Circle((0, 0), radius = 1, edgecolor = "black", fill = False)
    ax.add_patch(circ)

    tri = plt.Polygon(xy, color = "red", alpha = 0.8)
    ax.add_patch(tri)

    # Set axis limits.
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)

    # Add lines for the x and y axes.
    ax.axhline(color = "black", alpha = 0.2)
    ax.axvline(color = "black", alpha = 0.2)

    # Add faded gridlines.
    plt.grid(alpha = 0.1)

    # Show plot.
    plt.show()
```

```python
# Show triangle in a unit circle.    
unit_circle([
                [0, 0],
                [math.cos(math.pi / 3), math.sin(math.pi / 3)],
                [math.cos(math.pi / 3), 0]
            ])
```

Using a unit circle, the hypotenuse ($h_h$) is equal to the radius. The radius is equal to one because we are using a unit circle to find the side-ratios of the triangle. Since we know the hypotenuse, we can find the exact value of adjacent ($h_a$) side:

$$
cos(\pi / 3) = \frac{h_a}{h_h} \\
\Rightarrow cos(\pi / 3) = \frac{h_a}{1} \\
\Rightarrow h_a = cos(\pi / 3) \\
\Rightarrow h_a = \frac{1}{2} 
$$

To find the exact value for the opposite side ($h_o$), we can use Pythagoras' theorem:

$$
h_h^2 = h_a^2 + h_o^2 \\ 
\Rightarrow h_o^2 = h_h^2 - h_a^2  \\ 
\Rightarrow h_o^2 = (1)^2 - \left(\frac{1}{2}\right)^2  \\ 
\Rightarrow h_o^2 = 1 - \frac{1}{4}  \\
\Rightarrow h_o^2 = \frac{3}{4}  \\
\Rightarrow h_o = \frac{\sqrt{3}}{\sqrt{4}}  \\
\Rightarrow h_o = \frac{\sqrt{3}}{2}
$$

Finally, we can find the exact value of the tangent:

$$
tan(\pi / 3) = \frac{h_o}{h_a} \\
\Rightarrow tan(\pi / 3) = \frac{\left(\frac{\sqrt{3}}{2}\right)}{(1)} \\
\Rightarrow tan(\pi / 3) = \frac{\sqrt{3}}{2}
$$
